1. LASSO 回歸可以被用來作為 Feature selection 的工具，請了解 LASSO 模型為什麼可用來作 Feature selection  
A: LASSO的目標涵數是非光滑的。對於非光滑的最佳化問題，其最佳解可能會在導數為0處或者在不可導的地方，也就是各個角上。  
   對於多维的LASSO，所谓的"角"，就是那些很多特徵的係數為0的地方。   
   所以LASSO會给出一个稀疏解，能有特徵選擇的作用。
